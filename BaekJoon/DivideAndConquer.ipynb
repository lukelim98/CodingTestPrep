### 행렬 곱셈 (2740)

**Question**

N*M크기의 행렬 A와 M*K크기의 행렬 B가 주어졌을 때, 두 행렬을 곱하는 프로그램을 작성하시오.

---

**Input**

첫째 줄에 행렬 A의 크기 N 과 M이 주어진다. 둘째 줄부터 N개의 줄에 행렬 A의 원소 M개가 순서대로 주어진다. 그 다음 줄에는 행렬 B의 크기 M과 K가 주어진다. 이어서 M개의 줄에 행렬 B의 원소 K개가 차레대로 주어진다. N 과 M, 그리고 K는 100보다 작거나 같고, 행렬의 원소는 절대값이 100보다 작거나 같은 정수이다.

---

**Output**

첫째 줄부터 N개의 줄에 행렬 A와 B를 곱한 행렬을 출력한다. 행렬의 각 원소는 공백으로 구분한다.

```python
def matrix_multiplication():
    first_matrix = []
    second_matrix = []
    N, M = map(int, input().split())
    for _ in range(N):
        first_matrix.append(list(map(int, input().split())))
    M, K = map(int, input().split())
    for _ in range(M):
        second_matrix.append(list(map(int, input().split())))
    result = []
    for i in range(N):
        row = []
        for j in range(K):
            col_j = [second_matrix[r][j] for r in range(M)]
            val = sum(first_matrix[i][k] * col_j[k] for k in range(M))
            row.append(val)
        result.append(row)
    return result
    
result = matrix_multiplication()
for item in result:
    print(' '.join(map(str, item)))
```

### 쿼드트리 (1992)

**Question**

흑백 영상을 압축하여 표현하는 데이터 구조로 쿼드 트리(Quad Tree)라는 방법이 있다. 흰 점을 나타내는 0과 검은 점을 나타내는 1로만 이루어진 영상(2차원 배열)에서 같은 숫자의 점들이 한 곳에 많이 몰려있으면, 퀴드 트리에서는 이를 압축하여 간단히 표현할 수 있다.

주어진 영상이 모두 0으로만 되어 있으면 압축 결과는 "0"이 되고, 모두 1로만 되어 있으면 압축 결과는 "1"이 된다. 만약 0과 1이 섞여 있으면 전체를 한 번에 나타내지를 못하고, 왼쪽 위, 오른쪽 위, 왼쪽 아래, 오른쪽 아래, 이렇개 4개의 영상으로 나누어 압축하게 되면, 이 4개의 영역을 압축한 결과를 차례대로 괄오 안에 묶어서 표현한다.

![1922](../Images/1992.png)

위 그림에서 왼쪽의 영상은 오른쪽의 배열과 같이 숫자로 주어지며, 이 영상을 쿼드 트리 구조를 이용하여 압축하면 "(0(0011)(0(0111)01)1)"로 표현한다. N x N 크기의 영상이 주어질 때, 이 영상을 압축한 결과를 출력하는 프로그램을 작성하시오.

---

**Input**

첫째 줄에는 영상의 크기를 나타내는 숫자 N이 주어진다. N은 언제나 2의 제곱수로 주어지며, 1$\leq$N$\leq$64의 범위를 가진다. 두 번째 줄부터는 길이 N의 문자열이 N개 들어온다. 각 문자열은 0또는 1의 숫자로 이루어져 있으며, 영상의 각 점들을 나타낸다.

---

**Output**

여상을 압축한 결과를 출력한다.

```python
def quad_tree():
    def uniform(matrix):
        # Get the first element in the matrix
        first_element = matrix[0][0]
        for row in matrix:
            for element in row:
                if element != first_element:
                    return False
        return True

    def _recursion_quad_tree(matrix):
        if uniform(matrix):
            result.append(str(matrix[0][0]))
            return 
        else:  # divide the matrix into four different parts and run the recursion
            N = int(len(matrix)/2)
            top_left = [matrix[i][:N] for i in range(N)]
            top_right = [matrix[i][N:] for i in range(N)]
            bottom_left = [matrix[i][:N] for i in range(N, int(N*2))]
            bottom_right = [matrix[i][N:] for i in range(N, int(N*2))]
            result.append('(')
            _recursion_quad_tree(top_left)
            _recursion_quad_tree(top_right)
            _recursion_quad_tree(bottom_left)
            _recursion_quad_tree(bottom_right)
            result.append(')')
            
    N = int(input())
    matrix = []
    for _ in range(N):
        matrix.append(list(map(int, list(input()))))
    result = []
    _recursion_quad_tree(matrix)
    return result

    
result = quad_tree()
print(''.join(result))
```
